```python
import pickle5 as pickle
with open('maps_sample.pkl', 'rb') as h: M = pickle.load(h)
i2e, i2r = M['i2e'], M['i2r']
```

# load splits

```python
with open("train.txt") as f: lines = f.readlines()
with open("valid.txt") as f: lines += f.readlines()
with open("aux.txt") as f: lines += f.readlines()
with open("test.txt") as f: lines += f.readlines()
```

# create maps

```python
I2R, I2E = {}, {}
for line in lines:
    split = line.split("\t")
    r = split[0].strip()
    I2R[r] = i2r[int(r)] 
    
    for e in split[1:]: 
        e = e.strip()
        I2E[e] = i2e[int(e)]
```

# store raw names

```python
for line in lines:
    split = line.split("\t")
    r = split[0].strip()
    print(I2R[r]) 
    
    for e in split[1:]: 
        e = e.strip()
        print(I2E[e])
    print("\n\n\n")
```

```python
Raw = {} 
I2R
```

```python
r = '___relative_(subject)___relative_(object)___type_of_kinship___'
Raw[r] = 'Person, Relative, Kinship'
Raw
```

```python
i2raw = {}
for i in I2R:
    i2raw[i] = Raw[I2R[i]].replace(", ", "_")
```

```python
with open("relations.txt", "w") as f: 
    for i in i2raw: f.write(i + "\t" + i2raw[i] + "\n")
```
